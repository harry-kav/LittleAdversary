### Downloading and Preprocessing the Dataset

The functions used here are described in more detail in utils_data.ipynb, but have been shown here in a streamlined fashion to show the pipeline of data pre-processing.

```python
DATASET_PATH = '../data/mini_speech_commands' # get the dataset

data_dir = pathlib.Path(DATASET_PATH)
if not data_dir.exists():
  tf.keras.utils.get_file(
      'mini_speech_commands.zip',
      origin="http://storage.googleapis.com/download.tensorflow.org/data/mini_speech_commands.zip",
      extract=True,
      cache_dir='.', cache_subdir='data')
```

```python
# these are the command labels in the dataset
commands = np.array(tf.io.gfile.listdir(str(data_dir)))
commands = commands[(commands != 'README.md') & (commands != '.DS_Store')]
print('Commands:', commands)
```

```python
# here, we see how many files there are
filenames = tf.io.gfile.glob(str(data_dir) + '/*/*')
filenames = tf.random.shuffle(filenames)
num_samples = len(filenames)
print('Number of total examples:', num_samples)
print('Number of examples per label:',
      len(tf.io.gfile.listdir(str(data_dir/commands[0]))))
print('Example file tensor:', filenames[0])
```

```python
# create a train-test-validation split
train_files = filenames[:6400]
val_files = filenames[6400: 6400 + 800]
test_files = filenames[-800:]

print('Training set size', len(train_files))
print('Validation set size', len(val_files))
print('Test set size', len(test_files))
```

```python
def decode_audio(audio_binary): # decode the audio into a float tensor
  audio, _ = tf.audio.decode_wav(audio_binary)
  return tf.squeeze(audio, axis=-1)
```

```python
def get_label(file_path): # get the label of a given file
  parts = tf.strings.split(file_path, os.path.sep)

  return parts[-2] 
```

```python
def get_waveform_and_label(file_path): #get the audio waveform and label of a file
  label = get_label(file_path)
  audio_binary = tf.io.read_file(file_path)
  waveform = decode_audio(audio_binary)
  return waveform, label
```

```python
# map the data as a tensorflow dataset

AUTOTUNE = tf.data.experimental.AUTOTUNE
files_ds = tf.data.Dataset.from_tensor_slices(train_files)
waveform_ds = files_ds.map(get_waveform_and_label, num_parallel_calls=AUTOTUNE)
```

```python
# here, we visualise the waveforms
rows = 3
cols = 3
n = rows*cols
fig, axes = plt.subplots(rows, cols, figsize=(10, 12))
for i, (audio, label) in enumerate(waveform_ds.take(n)):
  r = i // cols
  c = i % cols
  ax = axes[r][c]
  ax.plot(audio.numpy())
  ax.set_yticks(np.arange(-1.2, 1.2, 0.2))
  label = label.numpy().decode('utf-8')
  ax.set_title(label)

plt.show()
```

```python
def get_spectrogram(waveform): #convert the waveforms into spectrogram images
    
  # Padding for files with less than 16000 samples
  zero_padding = tf.zeros([16000] - tf.shape(waveform), dtype=tf.float32)

  # Concatenate audio with padding so that all audio clips will be of the 
  # same length
  waveform = tf.cast(waveform, tf.float32)
  equal_length = tf.concat([waveform, zero_padding], 0)
  spectrogram = tf.signal.stft(
      equal_length, frame_length=255, frame_step=128)
      
  spectrogram = tf.abs(spectrogram)

  return spectrogram
```

```python
for waveform, label in waveform_ds.take(1): # display an audio file with label and spectrogram shape
  label = label.numpy().decode('utf-8')
  spectrogram = get_spectrogram(waveform)

print('Label:', label)
print('Waveform shape:', waveform.shape)
print('Spectrogram shape:', spectrogram.shape)
print('Audio playback')
display.display(display.Audio(waveform, rate=16000))
```

```python
def plot_spectrogram(spectrogram, ax): #visualise spectrogram 
  # Convert to frequencies to log scale and transpose so that the time is
  # represented in the x-axis (columns).
  log_spec = np.log(spectrogram.T)
  height = log_spec.shape[0]
  X = np.arange(16000, step=height + 1)
  Y = range(height)
  ax.pcolormesh(X, Y, log_spec)


fig, axes = plt.subplots(2, figsize=(12, 8))
timescale = np.arange(waveform.shape[0])
axes[0].plot(timescale, waveform.numpy())
axes[0].set_title('Waveform')
axes[0].set_xlim([0, 16000])
plot_spectrogram(spectrogram.numpy(), axes[1])
axes[1].set_title('Spectrogram')
plt.show()
```

```python
def get_spectrogram_and_label_id(audio, label): #get spectrogram image and label id
  spectrogram = get_spectrogram(audio)
  spectrogram = tf.expand_dims(spectrogram, -1)
  label_id = tf.argmax(label == commands)
  return spectrogram, label_id
```

```python
#get an image dataset of spectogram images and the speech labels
spectrogram_ds = waveform_ds.map(
    get_spectrogram_and_label_id, num_parallel_calls=AUTOTUNE)
```

```python
# visualise our spectrogram image dataset
rows = 3
cols = 3
n = rows*cols
fig, axes = plt.subplots(rows, cols, figsize=(10, 10))
for i, (spectrogram, label_id) in enumerate(spectrogram_ds.take(n)):
  r = i // cols
  c = i % cols
  ax = axes[r][c]
  plot_spectrogram(np.squeeze(spectrogram.numpy()), ax)
  ax.set_title(commands[label_id.numpy()])
  ax.axis('off')
  
plt.show()
```

```python
#function to do all the preprocessing steps that we have just outlined
def preprocess_dataset(files):
  files_ds = tf.data.Dataset.from_tensor_slices(files)
  output_ds = files_ds.map(get_waveform_and_label, num_parallel_calls=AUTOTUNE)
  output_ds = output_ds.map(
      get_spectrogram_and_label_id,  num_parallel_calls=AUTOTUNE)
  return output_ds
```

```python
#get our 3 data sets for training, validation and testing
train_ds = spectrogram_ds
val_ds = preprocess_dataset(val_files)
test_ds = preprocess_dataset(test_files)
```

```python
#create batches for training and validation
batch_size = 64
train_ds = train_ds.batch(batch_size)
val_ds = val_ds.batch(batch_size)
```

```python
#cache and prefetch to reduce latency during training
train_ds = train_ds.cache().prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)
```

### Building the model

Model has 16000 input nodes, which are resized to 1024- attempts to shrink any further lost too much accuracy.  
Two convolution layers, one with 32 filters, one with 64- both with 3x3 kernels/filter masks.  
One max pooling layer to downsample and retain the most prominent features of the images.  
A droput layer to reduce overfitting.  
A layer to flatten the dimensions of the input.  
A dense, hidden layer with 128 nodes with a relu activation function.  
Another dropout layer.  
The output layer with 8 nodes.  

```python
for spectrogram, _ in spectrogram_ds.take(1):
  input_shape = spectrogram.shape
print('Input shape:', input_shape)
num_labels = len(commands)

norm_layer = preprocessing.Normalization()
norm_layer.adapt(spectrogram_ds.map(lambda x, _: x))


model = models.Sequential([
    layers.Input(shape=input_shape), #input layer- 124x129 = 16000
    layers.Resizing(32, 32), #resize to be 32x32 = 1024
    norm_layer,
    layers.Conv2D(32, 3, activation='relu'), #convolution layers
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.25),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(num_labels), #output
])


model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'],
)
```

```python
EPOCHS = 10
history = model.fit(
    train_ds, 
    validation_data=val_ds,  
    epochs=EPOCHS,
    callbacks=tf.keras.callbacks.EarlyStopping(verbose=1, patience=2),
)
```

```python
metrics = history.history
plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.show()
```

### Testing the Model

```python
test_audio = []
test_labels = []

for audio, label in test_ds:
  test_audio.append(audio.numpy())
  test_labels.append(label.numpy())

test_audio = np.array(test_audio)
test_labels = np.array(test_labels)
```

```python
y_pred = np.argmax(model.predict(test_audio), axis=1)
y_true = test_labels

test_acc = sum(y_pred == y_true) / len(y_true)
print(f'Test set accuracy: {test_acc:.0%}')
```

### Confusion Matrix

The following confusion matrix shows us that our small CNN model is accurate for the majority of predictions for each class, although it struggles to differentiate between words with similar vowel sounds, like 'go' and 'no', whereas it found the more distinct sounds such as 'right' very easy to differentiate from the other speech commands.

```python
confusion_mtx = tf.math.confusion_matrix(y_true, y_pred) 
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_mtx, xticklabels=commands, yticklabels=commands, 
            annot=True, fmt='g')
plt.xlabel('Prediction')
plt.ylabel('Label')
plt.show()
```

### Confidence in predictions

The bar plot below reinforces the conclusions made from the confusion matrix, as we see that, when we input the speech command 'no', the neural network correctly predicts the 'no' class, but also has a low level of confidence that the command could be for 'go' or 'down'.

### Adding the ability to test with any wav file

```python
#to get this to work with any audio input, we use the wave module, which will replace decode_wav
import wave
obj = wave.open(str(sample_file), 'rb')
n_samples = obj.getnframes()
signal_wave = obj.readframes(n_samples)
signal_array = np.frombuffer(signal_wave, dtype=np.int16)
obj.close()

print(signal_array.shape)
```

```python
waveform = signal_array / 32768 #same as tf decode_wav
waveform = tf.convert_to_tensor(waveform, dtype = tf.float32)
spec = get_spectrogram(waveform)
spec = tf.expand_dims(spec,0)
prediction = loaded_model(spec)
print(prediction)

label_pred = np.argmax(prediction, axis=1) #get the label being predicted
print(commands[label_pred[0]])
plt.bar(commands, tf.nn.softmax(prediction[0]))
plt.title(f'Predictions for "{commands[label[0]]}"')
plt.show()
```
